### Review of the Abstract
>__Objective__:To assess the non-specific effect of Bacillus Calmette-Gurin (BCG) vaccination at birth on psychomotor development.

_This is a clinical trial study that investigates the non-specific effect of Bacillus Calmette-Gurin (BCG) vaccination on psychomotor development. By definition, Psychomotor development is changing in the cognitive, emotional, motor, and social capacities of a child from the beginning of life throughout_ 
- _fetal and neonatal periods,_ 
- _infancy,_ 
- _childhood, and_ 
- _adolescence_

_Looking at this objective, the initial thought here is that this clinical trial might requirer much time up to 19 years, which means following up the subjects until the end of their adolescence age.
Hence, this might involve missing data due to loss of interest in the subject (children might out-migrate).
Also, it would mean changes in condition and new agents (investigator or workers); which invites inconsistencies._

>__Design__:
This is a pre-specified secondary outcome from a randomized, clinical trial.

_This objective being a pre-specified secondary outcome in the protocol of another study of Calmette, it might have been easier to obtain ethical clearance approval. Also, this qualifies the procedure of randomized controlled trial (RCT), which guards against switching of outcomes._

>__Setting__:
Maternity units and paediatric wards at three university hospitals in Denmark.

_There are three centre (or hospitals), which might mean a multi-centre clinical trial. Since there might be differences in the staff and practices from one clinic to another, we would expect the analysis to adjust for these differences._

>__Participants__:
Children born at gestational age (GA) 32 weeks and above. All women planning to give birth at the three sites were invited during the recruitment period. Out of 4262 randomized children, 144 were premature (GA < 37 weeks). There were 2129 children (71 premature) randomized to BCG and 2133 randomized (73 premature) to the control group.

_This marks the inclusion and exclusion criteria and the sample size used for this study, which seems reasonable. Be that as may this criteria might not be limited to those mention in this section of the abstract. >__Interventions__:
>BCG vaccination 0.05 ml was given intradermally in the upper left arm at the hospital within seven days of birth. Children in the control group did not receive any intervention. Parents were not blinded to allocation.

_The intervention indicates further the practice of experimental study (here in the clinical trial), where one group received treatment, and another did not. However, the trial was not blinded. This might introduce bias to the in measurement, experiment or intervention._

_Given that the parents are the ones to provide the assessment the measurement are subjective and not standard, "Every mother want their child to be the best"._

> __Main outcome measures__:
Psychomotor development measured using the Ages and Stages Questionnaire (ASQ) completed by the parents at 12 months. Additionally, parents of premature children (gestational age < 37 weeks) completed an ASQ at 6 and 22 months. Developmental assessment was available for 3453/4262 (81%).

_We also notice about 19% of missingness; missing values are typical in clinical trials; hence, methods for dealing with the missingness should be readily available and specified in the protocol._

_This endpoint is subjective; therefore, it is often classified as the softest endpoint. Although the authors made an effort to scale this outcome, it is not clear what it is since the cognitive, emotional, motor, and social capacities might even need philosophers to define their meaning._

### Review of Blinding and Randomization
> __Trial methodology in brief__:
The trial took place at maternity units and paediatric wards at three university hospitals in Denmark. Within seven days of birth, study staff randomized children to receive BCG (Danish SSI strain 1331) in the standard dose of 0.05 mL intra-dermally on the left shoulder or to a no-intervention control group. Parents were not blinded to allocation since the local inflammatory reaction after BCG vaccination could not be mimicked. 

_The study identifies parents as potential subjects to be blinded; this was not done since it was not possible to mimic the local inflammatory reaction after BCG vaccination.

_Maybe agents could have been the ones to be blinded, which would mean they would be the ones to make observations on Psychomotor development. This would mean a standard way of measurement of the ASQ, thereby avoiding assessment bias._

> Randomization was stratified by prematurity (gestational age [GA] < 37 weeks or  37 weeks) and children were allocated 1:1 in permuting blocks of 2-4-6 using an online system.  

_Permuted block was good since it allowed the patients to be divided by whether they are pre-matures or not, thereby allowing a balanced group for the analysis of secondary outcome. However, no stratification by hospitals might be questionable. Also, permuting blocks randomization limit stratification by other potential prognostic factors (e.g. child sex, study site, maternal education, multiple births). This would in a way guarantees the distributional assumptions of the test statistics and estimators, which was one of the statistical issues that were experienced in this study; hence, non-parametric methods (Mann-Whitney U-test)._

_The 2-4-6 idea was also good because it avoids selection bias in the sense that the clinicians would not guess the next assignment for the patient._

_In a case of many prognostic factors, the permuted block might lead to fewer points in the permutation space. Hence, other randomization strategies might be encouraged._

The aspects of imbalances were revealed in the table that the authors included in the paper.
>

The table showed some slight imbalances within a few characteristics (or factors) for complete responders. The imbalances might have been slight because of the huge trial size. Otherwise, for small trials, these imbalances might have been worse.
Be that as may, for characteristics that may have been major prognosis, the imbalances may raise doubts about treatment effects.

Over and above, the randomization strategy used ensured balance within treatment groups and prematurity status; however, there where some accidental bais concerning some factors.

### Review of Distribution
>The mean ASQ score among the 3453 children with complete questionnaires at 12 months was 178.2 points (SD 52.4) in the BCG group vs. 179.9 points (SD 53.6) in the control group. A total of 22 children scored the maximum score of 300.

The authors showed a distribution of the BCG and control group of all the children at 12 months.
>

The distribution seems fairly normally distributed to perform a comparison using t-test for difference between means of the two groups since the variance of the shown to be approximately equal, as reported above.

We can also mimic the two distribution to see if they are comparable with what is displayed in the report.

```python
from matplotlib import pyplot as plt
%matplotlib inline
from scipy.stats import norm

import pandas as pd
import numpy as np
import random, uuid

random.seed(3658)
# Baseline variance between study arms
# Statistics for BCG
mu_BCG = 179.9 # 
sd_BCG = 53.6
x_BCG = np.arange(0, 300)
y_BCG = norm.pdf(x_BCG, mu_BCG, sd_BCG)
# Statistics for the control group
mu_control = 178.2
sd_control = 52.4
x_control = np.arange(0, 300)
y_control = norm.pdf(x_control, mu_control, sd_control)
# Draw the chart
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x_BCG,y_BCG, color="red", label="BCG")
ax.plot(x_control,y_control, color="blue", label="Control")
ax.legend(title='Groups',loc="upper left", frameon=True)
ax.set_title("Mimic distribution of the ASQ scores for the control and BCG groups", weight="bold")
for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(15)
plt.show()
```

### Concluding Remarks
Although soft-endpoint, such as in this study are appealing to the parents of the children in this study, soft-endpoint are not reliable unless they are double-blinded. Hence, the reliability of the ASQ scores provided by the mothers would have been justifiable if the was a way to blind them.

On the other hand, the assessment of the endpoint of unspecific-benefits might require a longer follow-up time.

Based on the results for the secondary outcome, it is not surprising that there were no conclusive results on effect since the hypotheses were powered at 50% but with a high effect size. The effect size was large, and it means that it was difficult to detect; hence it needed more power.

__Statistics:__
- The analysis treated the responses of the Questionnaire as separate assuming that there is no variation in the scores generated by the sub-domains (communication, fine motor development, gross motor development, problem-solving, and personal and social development). Perhaps the authors could have had a setup of a Multivariate analysis of some sort.
- As for non-normally distributed response Mann-Whitney U-test can be used for only two groups, perhaps authors could have a look at Kruskal Wallis test, which can compare two or more groups following the comment above.
- Handling of missing responses is not transparent. For example, it is not justifiable why the study assumed missing completely at random mechanism, let alone the sensitivity analysis for non-responses. Of course, there might be other ways of handling missing responses, but in this case, no reference is provided.

__Unadjusted Factors or Missed Opportunity:__
- Some studies showed that twins could cause parenting stress and depression compared to those with a single child. These aspects cause difficulties for families with twins, thereby contributing negatively towards the cognitive functioning of a child. 
- The study has not controlled for mothers who have been vaccinated for TB at their birth. This factor might be one of the confounding variables that can perhaps bias the results of this study.

__Acknolagement of Non-Generalisability:__
- The underlying effects of non-specific benefits speaks more volume towards the aspect of genetics. Given that the practice of BCG vaccination for TB is in low-mid income countries in southern Africa and Asia, the genetic formation for these populations varies significantly with the Danish one. Hence, this study might lack generalisability.
